## Step 2: Load the Dataset

```python
dataset_path = os.path.join('..', 'datasets', 'city_day_comprehensive_2026.csv')
print(f'Loading dataset from: {dataset_path}')
print(f'Current working directory: {os.getcwd()}')

if not os.path.exists(dataset_path):
    print(f'ERROR: File not found at {dataset_path}')
    print('Please ensure the dataset file exists in the datasets/ folder')
else:
    print('File found! Loading dataset...')
```

### Explanation of Commands:

- **os.path.join('..', 'datasets', 'filename.csv')**: Constructs the file path. The '..' means go up one directory level from notebooks/ to the main folder, then enter datasets/.

- **os.getcwd()**: Returns the current working directory (get current working directory) to help debug path issues.

- **os.path.exists()**: Checks whether the file exists at the specified path. Returns True if it exists, False otherwise.

```python
df = pd.read_csv(dataset_path, encoding='utf-8', low_memory=False)
print(f'Dataset loaded successfully!')
print(f'Shape: {df.shape[0]} rows, {df.shape[1]} columns')
print(f'Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB')
```

### Explanation of Commands:

- **pd.read_csv()**: Reads a CSV file and converts it into a pandas DataFrame (table-like structure).

- **encoding='utf-8'**: Specifies character encoding to handle special characters in city names.

- **low_memory=False**: Tells pandas to load entire columns at once to determine correct data types.

- **df.shape**: Returns (rows, columns) tuple showing DataFrame dimensions.

- **df.memory_usage(deep=True).sum()**: Calculates total memory used. Divide by 1024^2 to convert bytes to MB.

## Step 5: Remove Unnecessary Columns

```python
columns_to_drop = ['Unnamed: 0']
columns_to_drop = [col for col in columns_to_drop if col in df.columns]
print(f'Columns to remove: {columns_to_drop}')

if columns_to_drop:
    df.drop(columns=columns_to_drop, axis=1, inplace=True)
    print(f'Removed {len(columns_to_drop)} column(s)')
else:
    print('No columns to remove')
```

### Explanation:

- **columns_to_drop = ['Unnamed: 0']**: List of columns to remove (usually old index columns).

- **df.drop()**: Removes specified columns from DataFrame.

- **axis=1**: Indicates we are dropping columns (axis=0 would drop rows).

- **inplace=True**: Modifies the original DataFrame directly without creating a copy.

## Step 6: Convert Datetime Column

```python
df['Datetime'] = pd.to_datetime(df['Datetime'], format='%Y-%m-%d', errors='coerce')
print(f'Datetime converted! New type: {df["Datetime"].dtype}')
print(f'Invalid dates: {df["Datetime"].isna().sum()}')
```

### Explanation:

- **pd.to_datetime()**: Converts string dates to datetime objects for time series analysis.

- **format='%Y-%m-%d'**: Specifies date format (Year-Month-Day, e.g., 2020-11-09).

- **errors='coerce'**: Converts invalid dates to NaT (Not a Time) instead of raising errors.

## Step 7: Check for Duplicates

```python
duplicate_count = df.duplicated().sum()
print(f'Duplicate rows: {duplicate_count}')
print(f'Percentage: {(duplicate_count / len(df)) * 100:.2f}%')
```

### Explanation:

- **df.duplicated()**: Returns True/False for each row indicating if it is a duplicate.

- **.sum()**: Counts True values to get total duplicate rows.

```python
if duplicate_count > 0:
    df.drop_duplicates(keep='first', inplace=True)
    print(f'Duplicates removed! New shape: {df.shape}')
else:
    print('No duplicates found.')
```

### Explanation:

- **df.drop_duplicates()**: Removes duplicate rows, keeping only unique records.

- **keep='first'**: Keeps the first occurrence and removes subsequent duplicates.

## Step 8: Missing Value Analysis

```python
missing_values = df.isna().sum()
missing_df = pd.DataFrame({
    'Missing_Count': missing_values,
    'Missing_Percentage': (missing_values / len(df)) * 100
})
missing_df = missing_df.sort_values('Missing_Percentage', ascending=False)
print('Missing Values:')
print(missing_df[missing_df['Missing_Count'] > 0])
```

### Explanation:

- **df.isna()**: Returns True for missing values, False for present values.

- **.sum()**: Counts missing values per column.

- **sort_values()**: Sorts DataFrame by specified column.

- **ascending=False**: Sorts in descending order (highest first).

```python
plt.figure(figsize=(14, 7))
missing_cols = missing_df[missing_df['Missing_Count'] > 0].index
missing_pct = missing_df[missing_df['Missing_Count'] > 0]['Missing_Percentage']
bars = plt.barh(missing_cols[::-1], missing_pct[::-1], color='salmon', edgecolor='black')
plt.xlabel('Missing Values (%)', fontsize=12, fontweight='bold')
plt.ylabel('Columns', fontsize=12, fontweight='bold')
plt.title('Missing Values by Column', fontsize=14, fontweight='bold')
for bar, value in zip(bars, missing_pct[::-1]):
    plt.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height()/2, 
             f'{value:.1f}%', va='center', fontsize=10)
plt.tight_layout()
plt.show()
```

### Explanation:

- **plt.figure(figsize=(14, 7))**: Creates a new plot with specified width and height.

- **plt.barh()**: Creates horizontal bar chart (better for long labels).

- **[::-1]**: Reverses the list to display bars from top to bottom.

- **plt.text()**: Adds text annotations to the plot.

- **plt.tight_layout()**: Adjusts spacing to prevent labels from being cut off.

- **plt.show()**: Displays the plot.

## Step 10: Categorical Analysis

```python
categorical_cols = ['City', 'State', 'Region', 'AQI_Bucket']
for col in categorical_cols:
    print(f'\n{col}:')
    print(f'  Unique values: {df[col].nunique()}')
    print(f'  Top 10:')
    print(df[col].value_counts().head(10))
    print('-' * 50)
```

### Explanation:

- **df[col].nunique()**: Counts unique values in a column.

- **df[col].value_counts()**: Returns frequency count of each unique value.

- **.head(10)**: Shows only top 10 most common values.

## Step 11: City-wise Analysis

```python
city_stats = df.groupby('City').agg(
    Record_Count=('Datetime', 'count'),
    Start_Date=('Datetime', 'min'),
    End_Date=('Datetime', 'max'),
    AQI_Mean=('AQI', 'mean'),
    AQI_Max=('AQI', 'max')
).reset_index()
city_stats['Time_Span_Days'] = (city_stats['End_Date'] - city_stats['Start_Date']).dt.days
print('City Statistics (Top 20):')
city_stats.sort_values('Record_Count', ascending=False).head(20)
```

### Explanation:

- **df.groupby('City')**: Groups data by city to calculate per-city statistics.

- **.agg()**: Performs multiple aggregation operations on grouped data.

- **.reset_index()**: Converts grouped result back to regular DataFrame.

- **.dt.days**: Extracts number of days from time difference.

## Step 12: AQI Distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
axes[0].hist(df['AQI'].dropna(), bins=100, color='orange', edgecolor='black', alpha=0.7)
axes[0].set_xlabel('AQI Value')
axes[0].set_ylabel('Frequency')
axes[0].set_title('AQI Distribution')
axes[1].boxplot(df['AQI'].dropna(), vert=False)
axes[1].set_xlabel('AQI Value')
axes[1].set_title('AQI Box Plot')
plt.tight_layout()
plt.show()
```

### Explanation:

- **plt.subplots(1, 2)**: Creates figure with 1 row and 2 columns of plots.

- **axes[0]**: Accesses first subplot (left side).

- **.dropna()**: Removes missing values before plotting.

- **axes[1].boxplot()**: Creates box plot showing median, quartiles, and outliers.

## Step 13: AQI Bucket Distribution

```python
bucket_counts = df['AQI_Bucket'].value_counts()
plt.figure(figsize=(10, 8))
colors_dict = {'Good': '#22C55E', 'Satisfactory': '#84CC16', 'Moderate': '#F59E0B', 
               'Poor': '#F97316', 'Very Poor': '#EF4444', 'Severe': '#8B0000'}
bucket_colors = [colors_dict.get(b, '#808080') for b in bucket_counts.index]
plt.pie(bucket_counts.values, labels=bucket_counts.index, autopct='%.1f%%',
        colors=bucket_colors, startangle=90, explode=[0.02]*len(bucket_counts))
plt.title('AQI Category Distribution', fontsize=14, fontweight='bold')
plt.show()
```

### Explanation:

- **plt.pie()**: Creates pie chart showing category proportions.

- **autopct='%.1f%%'**: Displays percentages with 1 decimal place.

- **startangle=90**: Rotates chart to start from top.

- **explode=[0.02]**: Separates slices slightly for visibility.

## Step 14: Regional Comparison

```python
region_stats = df.groupby('Region').agg(
    Record_Count=('AQI', 'count'),
    AQI_Mean=('AQI', 'mean'),
    Cities_Count=('City', 'nunique')
).reset_index().sort_values('AQI_Mean', ascending=False)

plt.figure(figsize=(12, 6))
bars = plt.bar(region_stats['Region'], region_stats['AQI_Mean'], 
               color=['#EF4444', '#F97316', '#F59E0B', '#84CC16', '#22C55E'][:len(region_stats)],
               edgecolor='black', alpha=0.8)
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 2,
             f'{bar.get_height():.0f}', ha='center', fontsize=11)
plt.xlabel('Region')
plt.ylabel('Average AQI')
plt.title('Average AQI by Region')
plt.tight_layout()
plt.show()
```

### Explanation:

- **plt.bar()**: Creates vertical bar chart for regional comparison.

- **ha='center'**: Centers text horizontally over bars.
